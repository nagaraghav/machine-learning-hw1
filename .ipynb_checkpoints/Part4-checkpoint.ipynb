```python
import numpy as np
import pandas as  pd 
import matplotlib.pyplot as plt

```

```python
#Part 4
df = pd.read_table("auto-mpg.data" , header = None,sep='\s+' , names= 
    ["mpg" ,"cylinders", "displacement", "horsepower","weight",
                    "accelaration","model_year","origin","car_name"])
#remove incomplete data
df = df[df.horsepower != '?']
df = df.sample(frac=1).reset_index(drop=True)
```

```python
Y_mpg = df['mpg'].values
Y_mpg = Y_mpg.reshape(392,1)

Y_mpg_train = Y_mpg[0:292]
Y_mpg_test = Y_mpg[292:392]
```

```python
df = df.drop(['mpg', 'car_name'], axis=1) # (392, 7)
#df
```

```python
labels = df.columns # (7,)
```

```python
X = df.values # (392, 7)
X = X.astype('float64')
X
```

```python

def ols(input_f,output,degree):
    
    X = feature_expander(input_f, degree)

    W = np.dot(np.dot(np.linalg.pinv(np.dot(X.T, X)), X.T), output)
    return  W , X

def feature_expander(feature,degree):
    X = np.ones((feature.shape[0],1))
    for i in range(1, degree + 1):
        new_col = np.power(feature, i)
        X = np.concatenate((X, new_col), axis=1)
    return X

def MSE(predicted_values, output_values):
    num_samples = predicted_values.shape[0]
    return np.divide((np.sum(np.power((output_values - predicted_values),2))), num_samples)
```

```python
for i in range(X.shape[1]):
    cur_feature = X[:,i].reshape(392,1).astype('float64')
    cur_feature_train = cur_feature[0:292]
    cur_feature_test = cur_feature[292:392]
    
    X_newv = np.arange(min(cur_feature_test), max(cur_feature_test), ( max(cur_feature_test) - min(cur_feature_test))/10000)  
    X_newv = X_newv.reshape(X_newv.shape[0],1)
    plt.figure(figsize=(18, 10))
    plt.scatter(cur_feature_test,Y_mpg_test)

    
    for j in range(4):
        theta, ft_train_exp = ols(cur_feature_train, Y_mpg_train, j)
        hyp_train = np.dot(ft_train_exp, theta)
        mse_train = MSE(hyp_train, Y_mpg_train)
        print(labels[i] + " train data MSE " + str(j) + "-order " + ": " + str(mse_train))
        cur_test_ft_exp = feature_expander(cur_feature_test, j)
        
        hyp_test = np.dot(cur_test_ft_exp, theta)
        mse_test = MSE(hyp_test, Y_mpg_test)
        print(labels[i] + " test data MSE " + str(j) + "-order " + ": " + str(mse_test))
    
    
        X_new = feature_expander(X_newv, j)                           #Bump up features to plot
        plot_predictions = np.dot(X_new, theta)                      #Find predictions
        plt.plot(X_newv, plot_predictions, label = j)                              #Plot predictions
        plt.xlabel(labels[i])
        plt.ylabel("MPG")
        plt.legend()
        plt.title("Polynomial lines on testing data")
        
    #plt.show()
    print('\n')
```
