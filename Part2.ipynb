```python
import numpy as np
import pandas as  pd 
import matplotlib.pyplot as plt
import seaborn as sb
import import_ipynb
import Part1
```

```python
df = Part1.df
Y = df.mpg
Y = Y.values.reshape(392,1)
df.dtypes
```

```python
color = np.full((392,1), "very high")
for i in range(392):
    if Y[i] <= Part1.low:
        color[i] = "low"
    if Y[i] > Part1.low and Y[i] <= Part1.med:
        color[i] = "medium"
    if Y[i] > Part1.med and Y[i] <= Part1.high:
        color[i] = "high"
    if Y[i] > Part1.high:
        color[i] = "very high"
df["color"] = color 
```

```python
dfc = df
dfc = dfc.drop(['mpg'], axis=1)
df["color"].value_counts()
sb.set(style="ticks")
# passing in category for hue results in different color plots for each category
sb.pairplot(dfc, hue='color')
plt.show()
```
